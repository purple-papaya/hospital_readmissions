# Univariate Analysis

```python
for col in df.describe().columns:
    sns.histplot(df[col])
    plt.show()
```

```python
for col in df.describe().columns:
    sns.boxplot(df[col])
    plt.title(f'Boxplot of {col} values')
    plt.show()
```

```python
categorical_columns = [
    'age', 'medical_specialty', 
    'diag_1', 'diag_2', 'diag_3', 
    'glucose_test', 'A1Ctest', 'change', 
    'diabetes_med', 'readmitted'
    ]
```

```python
for col in categorical_columns:
    bar_df = df[[col]].value_counts().sort_values(ascending=False).reset_index()
    sns.barplot(bar_df, x=bar_df.iloc[:, 0], y=bar_df.iloc[:, 1])
    plt.title(f'Barplot of {col} values')
    plt.xticks(rotation=45)
    plt.show()
```
